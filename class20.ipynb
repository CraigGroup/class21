But, there is more to the data than than just the one-dimensional distributions.  What about correlations between variables?  To better understand correlations, we can make scatter plots of each variable pair.  Here, I will choose one pair of variables and make the scatter plot for Setosa:  sepal_length v/s sepal_width.

```python
plt.scatter(  
        np_list_virginica[:,0],        
        np_list_virginica[:,1],
        color='green',              
        label='virginica'   
)

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.title('Iris dataset: sepal length vs sepal width')
plt.legend(loc='lower right')

plt.show()
```

Now, we can see how the variables are correlated with each other.    In other words, it looks like there is some tendancy for the width the be smaller if the length is smaller.  This makes sense...

We can calculate Pearson's correlation coefficient to understad the level of correlation in the sample.  It looks slightly linear with a positive slope... 

```python
Pearson = np.corrcoef(np_list_virginica[:,0], np_list_virginica[:,1])
print(f'the correlation coefficient is \n {Pearson}')
```

Note the ones on the diagonals, indicating that the correlation coefficient of a variable with itself is one.  The offdiagonal is the coefficient of correlation between the two variables.
